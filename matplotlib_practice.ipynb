```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set y-axis to a list of floats as the total fare in US dollers accunulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
#fig, ax = plt.subplots()

#fig = plt.figure()
#ax = fig.add_subplot()

ax = plt.axes()

ax.plot(x_axis,y_axis)
```

```python
# Create the plot and add the label for the legend
plt.plot(x_axis,y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Create plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Add x and y axis labels
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add gridlines
plt.grid()

# Add legend
plt.legend()
```

```python
# Create plot using Object-Oriented Functions
fig, ax = plt.subplots()

# Color green, marker diamond
ax.plot(x_axis, y_axis, color='green', marker="d", linewidth=2, label='Boston')

# Add legend
ax.legend()

# Add title and labels
ax.set_title("Pyber Fares by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")

# Set y-axis limit between 0 and 45
ax.set_ylim(0,45)

# Add gridlines
ax.grid()
```

```python
# Create bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta",label='Boston')

# Add legend
plt.legend()

# Add title and axis labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")

# Invert y-axis so Jan is on top
plt.gca().invert_yaxis()
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create horizontal bar with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")

# Add Title and Labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")

plt.gca().invert_yaxis()
```
