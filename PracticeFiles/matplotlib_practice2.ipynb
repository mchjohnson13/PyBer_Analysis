```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set x-axis to strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dolors
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Get the standard deviation of values in y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create errorbar line chart with object oriented method
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create a bar plot with error bars
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add dependencies
from matplotlib.ticker import MultipleLocator
import numpy as np

# Create a horizontal bar chart and set the major tick marks on the x-axis
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()

plt.minorticks_on()
```
