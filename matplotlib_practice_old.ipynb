```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set y-axis to a list of floats as the total fare in US dollers accunulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()

ax.plot(x_axis,y_axis)
```

```python
# Create another plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis,y_axis)
```

```python
# Create a third plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```
