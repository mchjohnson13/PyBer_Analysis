```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set y-axis to a list of floats as the total fare in US dollers accunulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
#fig, ax = plt.subplots()

#fig = plt.figure()
#ax = fig.add_subplot()

ax = plt.axes()

ax.plot(x_axis,y_axis)
```

```python
# Create the plot and add the label for the legend
plt.plot(x_axis,y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Create plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Add x and y axis labels
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add gridlines
plt.grid()

# Add legend
plt.legend()
```

```python
# Create plot using Object-Oriented Functions
fig, ax = plt.subplots()

# Color green, marker diamond
ax.plot(x_axis, y_axis, color='green', marker="d", linewidth=2, label='Boston')

# Add legend
ax.legend()

# Add title and labels
ax.set_title("Pyber Fares by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")

# Set y-axis limit between 0 and 45
ax.set_ylim(0,45)

# Add gridlines
ax.grid()
```

```python
# Create bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta",label='Boston')

# Add legend
plt.legend()

# Add title and axis labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")

# Invert y-axis so Jan is on top
plt.gca().invert_yaxis()
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create horizontal bar with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")

# Add Title and Labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")

plt.gca().invert_yaxis()
```

```python
# Create a scatter plot
#plt.plot(x_axis, y_axis, 'o')
plt.scatter(y_axis, x_axis, color="red", label='Chicago')

# Add legend
plt.legend()

# Add title and labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")

# Add x-axis limit
plt.xlim(0,45)

# Invert y-axis so Jan is on top
plt.gca().invert_yaxis()
```

```python
# Create a bubble plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Increase values in y_axis by 3x
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)


# Create a bubble plot
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create bubble plot by increasing y_axis in plt.scatter()

plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# Create a scatter plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", edgecolors="black", linewidth=2, s=[i * 5 for i in y_axis], alpha=0.2, label='Boston')

# Add legend
ax.legend()

# Add title and axis labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")

# Set the x-axis limit between 0 and 45
ax.set_xlim(0,45)

# Invert the y axis so Jan is on top
plt.gca().invert_yaxis()
```

```python
# Explode month of July
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Assign colors for months
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Size chart
plt.subplots(figsize=(8, 8))

# Create a pie chart using plt method
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
```

```python
# Explode values
explode_2 = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Set colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Create pie plot using object oriented method
# Set size to 8x8
fig, ax = plt.subplots(figsize=(8, 8))

# Add percentage labels, change colors, explode Mar and Jul, change start angle so Jan is on top, add shadow
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', colors=colors, explode=explode_2, startangle=270, shadow=True)

# Set Jan-Dec to appear clockwise
plt.gca().invert_yaxis()
```
